# Index list
## Gets a list of AccuWeather indices

```python
import requests
import json
import pandas as pd
```

## Get api key

```python
from getpass import getpass
apikey = getpass('apikey')
```

## Get list

```python
index_list_url = 'http://dataservice.accuweather.com/indices/v1/daily'
query = dict(apikey=apikey)
indices = json.loads(requests.get(index_list_url, params=query).text)
```

```python
index_df = pd.DataFrame(indices)
```

```python
pd.set_option('display.max_rows', len(index_df))
```
